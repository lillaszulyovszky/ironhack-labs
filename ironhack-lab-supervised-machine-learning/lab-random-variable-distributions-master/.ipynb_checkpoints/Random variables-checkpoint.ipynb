```python
#separate numerical and categorical columns
df_nums = df._get_numeric_data()
df_cats = df.drop(columns=df_nums.columns)
```

```python
sns.distplot(df_nums['Customer Lifetime Value'])
plt.show()
```

```python
sns.distplot(df_nums['Income'])
plt.yscale('log')
plt.show()
```

```python
sns.distplot(df_nums['Total Claim Amount'])
plt.show()
```

```python
#function to rescale with log method 
def log_transfom_clean_(x):
    x = np.log(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN #  we can replace NANs with mean values in the next step   
df_nums['Customer Lifetime Value'] = list(map(log_transfom_clean_, df_nums['Customer Lifetime Value']))
```

```python
df_nums['Customer Lifetime Value'] = df_nums['Customer Lifetime Value'].fillna(np.mean(df_nums['Customer Lifetime Value']))
```

```python
#check how the distribution looks now
sns.distplot(df_nums['Customer Lifetime Value'])
plt.show()
```
